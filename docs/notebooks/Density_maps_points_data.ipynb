# Density maps

This tutorial shows how to create density maps for 3D point data (works analogously for 2D data). This function uses [vedo](https://vedo.embl.es/) to calculate density maps.

```python
import napari
import napari_spatial_statistics as nss
```

```python
viewer = napari.Viewer(ndisplay=3)

random_points = nss.make_random_points(n_classes=1, spatial_size=100)
viewer.add_points(random_points[0], **random_points[1])
napari.utils.nbscreenshot(viewer, canvas_only=False)
```

Now we can directly calculate a density map from this:

```python
density = nss.density_map(random_points[0], radius=15)
viewer.add_image(density, blending='additive', colormap='inferno')
napari.utils.nbscreenshot(viewer, canvas_only=False)
```
